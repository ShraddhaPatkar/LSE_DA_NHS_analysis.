# User defined function to segregate data with missing values
Function to find/segregate any unmapped/unknown values from any dataframe column, this function support dynamic parameters

```python
# user-defined function to find rows that contain unmapped or unknown values
def df_contains_word(df,column_name, column_value):
    """ does the dataframe column contain unmapped or unknown values? """
    y = df[df[column_name].str.contains(column_value, case=False)] 
    return y

# Multiple function call.
print("Rows from dataframe nc which contain unmapped values :")
unmapped_nc = df_contains_word(nc,'service_setting','Unmapped')
unmapped_nc
```

```python
# Assign empty list to new variable
tags = list()

# Retrieve hashtags from tweets
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
            
#Pandas Series to count the values in the list            
tags_series = pd.Series(tags)
tags_counts = tags_series.value_counts()
tags_counts.head(30)
```

## Q.Should the NHS start looking at increasing staff levels?

```python
#subset the data frame with appointment details
ar_agg = ar[['appointment_month', 'hcp_type','appointment_status',
             'appointment_mode','time_between_book_and_appointment','count_of_appointments']]
#calculate the count_of_appointments for different healthcare professionals permonth 
ar_agg_hcp = ar_agg.groupby(['hcp_type','appointment_month']).agg({'count_of_appointments':sum}).reset_index()
ar_agg_hcp
```

```python
# Create a grouped barplot.
plt.figure(figsize=(15, 12))
plot_ar=ar_agg_hcp[(ar_agg_hcp['appointment_month']>'2021-08') ]
sns.barplot(x='appointment_month', y='count_of_appointments', hue='hcp_type', data=plot_ar)

#improve readability of graph
plt.xlabel("Month of appointment")
plt.ylabel("Number of Appointments")
plt.title("How do the healthcare professional types differ over time?")
```

```python
#total number of appointments per month. 
ar_df = ar.groupby(['appointment_month']).agg({'count_of_appointments':sum}).reset_index()
#average utilisation of services:
ar_df['utilisation'] = ar_df['count_of_appointments'] / 30
ar_df['utilisation']=ar_df['utilisation'].round(decimals = 1)

#The NHS can accommodate a maximum of 1,200,000 appointments per day.
#Calculate % utilisation per month based on this
ar_df['utilisation_percent'] = (ar_df['utilisation'] * 100)/1200000
ar_df['utilisation_percent']=ar_df['utilisation_percent'].round(decimals = 1)
ar_df
```

```python
#point plot for % utilisation every month
plt.figure(figsize=(15, 12))
sns.pointplot(x='appointment_month', y='utilisation_percent', data=ar_df)
plt.xticks(rotation=45)
```

```python
#monthly attended appointments use merge to add attended appointments
ar_attended = ar.groupby(['appointment_month', 'appointment_status']).agg({'count_of_appointments':sum}).reset_index()
ar_attended = ar_attended[ar_attended['appointment_status'] == 'Attended']
ar_attended
# Use the left join to merge the two DataFrames.
attended_utilisation = pd.merge(ar_df, ar_attended,
                                    on='appointment_month', how='left')

attended_utilisation['Attended_utilisation'] = attended_utilisation['count_of_appointments_y'] / 30
attended_utilisation['Attended_utilisation']= attended_utilisation['Attended_utilisation'].round(decimals = 1)

#The NHS can accommodate a maximum of 1,200,000 appointments per day.
#Calculate % utilisation per month based on this
attended_utilisation['Attended_percent'] = (attended_utilisation['Attended_utilisation'] * 100)/1200000
attended_utilisation['Attended_percent']=attended_utilisation['Attended_percent'].round(decimals = 1)

attended_utilisation
```

```python
# Create an empty plot. Matplotlib
fig, ax_1 = plt.subplots()
ax_2 = ax_1.twinx()
ax_1.plot(attended_utilisation['appointment_month'], attended_utilisation['Attended_percent'], 'g-')
ax_2.plot(attended_utilisation['appointment_month'], attended_utilisation['utilisation_percent'], 'b-')
ax_1.set_xlabel('Month of appointment')
ax_1.set_ylabel('Actual Utilisation', color='g')
ax_2.set_ylabel('Utlisation as per bookeed appointments', color='b')
```
